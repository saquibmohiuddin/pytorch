## Tensor Data Types
*  The PyTorch uses `torch.float32` as default data type and switches to this dtype if it faces issues casting the elements of the tensor to the explicitly provided dtype.
This mainly due to computer science in precision or single precision. 

**It is important to ensure your data types and shape of the tensors as they can cause common errors in Machine Learning**

There are quite few important functions which include:
- a) `dtype` = Used to specify data type
- b) `device` = what device is the tensor on - can be used to specify "CPU" or "GPU" (`CUDA`)
- c) `requires_grad` = can be used to track gradients via pytorch

```python
tensor_float16 = torch.tensor([1, 2, 3], dtype=torch.float16, device=None, requires_grad=False)
tensor_float16
```

### Tensor Attributes

```python
tensor_float16.dtype
```

```python
tensor_float16.shape
```

```python
tensor_float16.device
```

## Tensor Operations - Manipulating Tensors
A myriad of operations can be performed on tensors which can be arithemetic or logical

```python
tensor_a = torch.tensor([1, 2, 3])
tensor_a
```

```python
print(tensor_a + 10)
print(tensor_a * 10)
print(tensor_a - 2)
```

### PyTorch Built-in Functions
PyTorch provides myriad of built-in functions which includes
* `torch.mul`
* `torch.add`

```python
tensor_b = torch.mul(tensor_a, 10)
tensor_b
```
