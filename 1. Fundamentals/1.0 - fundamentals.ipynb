### Random Tensors

* Neural Networks require weights to be initilaised randomly, and then update weights using various techniques to optimise those weights, hence random tensors are of important value.

```python
random_tensor = torch.rand(3, 4, 2)
random_tensor
```

```python
random_tensor.shape
```

```python
random_tensor.ndim
```

```python
# random image tensor
random_image_tensor = torch.rand(size=(224, 224, 3))
random_image_tensor
```
